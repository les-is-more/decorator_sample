### Goal

Let us build a data quality analyzer which will measure each columns based on
<ol><li> Data/Row Redundancy
    <li> Data Completeness vs Total Rows possible
    <li> Type Check
    <li> Encoding Check
    </ol>

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, Imputer, add_dummy_feature, LabelBinarizer, OneHotEncoder, PolynomialFeatures
from sklearn.feature_extraction import DictVectorizer, FeatureHasher
from sklearn import datasets
import numpy as np
```

```python
# for plotting the correlation heatmap, plus setting the font family for the 
import seaborn as sns
import matplotlib.pyplot as plt

font = {'family' : 'Verdana',
        'weight' : 'normal',
        'size'   : 22}

plt.rc('font', **font)
%matplotlib inline
```

```python
breastC_glob = datasets.load_breast_cancer()
breast = pd.DataFrame(breastC_glob.data, 
                      columns = breastC_glob['feature_names'])

# selecting only the columns with 'mean' prefix

corr_threshold = 0.0

bcorr = breast[breast.columns[breast.columns.str.contains('mean')]].corr()
bcorr = bcorr[abs(bcorr) > corr_threshold]

cols = bcorr.columns
ind = bcorr.index

# tedious job of converting from pandas to numpy array then to pandas dataframe again
bcorr_temp = np.triu(bcorr.values, k=0)
bcorr_temp[np.tril_indices(bcorr_temp.shape[0], 0)] = np.nan

bcorr2 = pd.DataFrame(bcorr_temp, columns=cols, index=ind)
print("Lowest correlation in the dataset: {:.2f}".format(bcorr.min(axis=0).min()))

plt.figure(figsize=(9,8))
plt.title("Summary of Variable Relationships")


sns.heatmap(bcorr2, cmap='icefire_r', vmax=1.0, vmin=-1.0,
            linewidths=0.2, annot=True, annot_kws={'size':8}, square=True)
```

```python
breastDF = pd.DataFrame(breastC_glob.data, columns = breastC_glob.feature_names)
breastDF['status'] = pd.Categorical.from_codes(breastC_glob.target, breastC_glob.target_names)
temp = breastDF[[*breastDF.columns[breastDF.columns.str.contains('mean')],'status']]

# g = sns.FacetGrid(breastDF, hue="status")
# g.map(plt.hist, 'mean area');
# g.map_diag(plt.hist)
# g.map_offdiag(plt.scatter)
# g.add_legend()

g = sns.PairGrid(temp, hue="status")
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
g.add_legend()
g.savefig('breast_cancer.png')
```

# Modelling Part

```python
# Model the breast cancer using Supper Vector Machine Classifier
from sklearn.model_selection import train_test_split
from sklearn import metrics

Y = breastC_glob['target']
X = breast[breast.columns[breast.columns.str.contains('mean')]]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 1)
```

```python
from sklearn import svm

svc_fit = svm.SVC(gamma='scale')
svc_fit.fit(X_train, Y_train)

Y_pred = svc_fit.predict(X_test)
print("Accuracy of the Support Vector Machine model: {:.2f}% ".format(metrics.accuracy_score(Y_test, Y_pred) * 100))
```

```python
Y_pred_train = svc_fit.predict(X_train)

svm_overfit = metrics.accuracy_score(Y_train, Y_pred_train) - metrics.accuracy_score(Y_test, Y_pred)

print("The overfit is at {:.2f}%".format(svm_overfit * 100))
```

```python
from  sklearn.tree import DecisionTreeClassifier

# random_state specifies the random seed
DT_clf = DecisionTreeClassifier(random_state = 1990)
DT_clf.fit(X_train, Y_train)
Y_pred = DT_clf.predict(X_test)

print("Decision Tree Accuracy: {:.2f}%".format(metrics.accuracy_score(Y_test, Y_pred) * 100))
```

### Logistic Regression Modelling:
<ol><li> Subject the Decision Tree model to bootstrapping to get the distribution of accuracy scores.
    <li> Experiment with different types of solvers, especially those that computationally converges for your dataset

```python
from sklearn import linear_model as lm

LR_breast = lm.LogisticRegression(solver='liblinear')
LR_breast.fit(X_train, Y_train)
Y_pred = LR_breast.predict(X_test)

print("Logistic Regression Accuracy: {:.2f}%".format(metrics.accuracy_score(Y_test, Y_pred) * 100))
```
